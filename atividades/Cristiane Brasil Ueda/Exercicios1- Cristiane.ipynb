Nota: A matriz para a mudança de variáveis deve ser definida positiva.

$$
f(x_1, x_2) = x_1^2 + 2x_1x_2 + 2x_2^2, \quad \bar{x} = (1, 1)^{\top}
$$


````python
import sympy as sp

# Definir variáveis simbólicas
x1, x2 = sp.symbols('x1 x2')
x = sp.Matrix([x1, x2])

# Definir a função
f = x1**2 + 2*x1*x2 + 2*x2**2

# 1. Gradiente
grad_f = sp.Matrix([sp.diff(f, var) for var in x])
print("Gradiente:")
sp.pprint(grad_f)

# 2. Hessiano
hessian_f = sp.hessian(f, x)
print("\nHessiano:")
sp.pprint(hessian_f)

# 3. Convexidade
eigenvals = hessian_f.eigenvals()
print("\nAutovalores do Hessiano (para testar convexidade):")
sp.pprint(eigenvals)

# 4. Curvatura na direção d no ponto x = (1, 1)
x_bar = {x1: 1, x2: 1}
d = sp.Matrix([1, 0])  # direção ao longo de x1

curvatura = (d.T * hessian_f.subs(x_bar) * d)[0]
print("\nCurvatura na direção d = [1, 0] no ponto x̄ = (1,1):")
sp.pprint(curvatura)

# 5. Mudança de variáveis - pré-condicionamento
H_at_xbar = hessian_f.subs(x_bar)
L = H_at_xbar.cholesky()
print("\nMatriz de pré-condicionamento (Cholesky do Hessiano):")
sp.pprint(L)
````

````python
import sympy as sp

# 1. Definir as variáveis e a função
x1, x2 = sp.symbols('x1 x2')
f = x1**2 + x2**2

# 2. Gradiente (∇f)
grad_f = [sp.diff(f, var) for var in (x1, x2)]
grad_f_vec = sp.Matrix(grad_f)

# 3. Hessiana (∇²f)
hessian_f = sp.hessian(f, (x1, x2))

# 4. Encontrar pontos críticos (onde gradiente é zero)
critical_points = sp.solve(grad_f, (x1, x2), dict=True)

# 5. Verificar condições de otimalidade
def is_positive_definite(matrix):
    return all(e > 0 for e in matrix.eigenvals())

print("Função objetivo: f(x1, x2) =", f)
print("\nGradiente:")
sp.pprint(grad_f_vec)

print("\nHessiana:")
sp.pprint(hessian_f)

print("\nPontos críticos encontrados:")
for point in critical_points:
    sp.pprint(point)

print("\nVerificação das condições de otimalidade:")

for i, point in enumerate(critical_points):
    subs_hessian = hessian_f.subs(point)
    print(f"\n Ponto Crítico {i+1}: {point}")
    print("Hessiana neste ponto:")
    sp.pprint(subs_hessian)
    
    if is_positive_definite(subs_hessian):
        print("A Hessiana é definida positiva. Mínimo local (e global).")
    else:
        print("A Hessiana não é definida positiva. Não é um mínimo.")

````

### Exercise 5.2

### Problema:
Minimizar a seguinte função objetivo:

$$
\min_{x \in \mathbb{R}^2} f(x) = \frac{1}{3}x_1^3 + x_2^3 - x_1 - x_2
$$


````python
import sympy as sp

# Variáveis simbólicas
x1, x2 = sp.symbols('x1 x2')

# Função objetivo
f = (1/3)*x1**3 + x2**3 - x1 - x2

# 1. Gradiente e Hessiana
grad_f = [sp.diff(f, var) for var in (x1, x2)]
hess_f = sp.hessian(f, (x1, x2))

print("Gradiente:")
sp.pprint(grad_f)

print("\nHessiana:")
sp.pprint(hess_f)

# 2. Pontos críticos: gradiente = 0
crit_points = sp.solve(grad_f, (x1, x2), dict=True)
print("\nPontos críticos:")
for pt in crit_points:
    print(pt)

# 3. Verificação das condições de otimalidade necessárias
print("\nVerificação da Hessiana em cada ponto crítico:")
for pt in crit_points:
    H = hess_f.subs(pt)
    eigenvals = H.eigenvals()
    print(f"\nPonto crítico: {pt}")
    print("Hessiana:")
    sp.pprint(H)
    print("Autovalores:", list(eigenvals.keys()))
    
    if all(val > 0 for val in eigenvals):
        print("Mínimo local (condição suficiente satisfeita)")
    elif all(val < 0 for val in eigenvals):
        print("Máximo local")
    else:
        print("⚠️ Ponto de sela (condição suficiente NÃO satisfeita)")

# Avaliação da função nos pontos críticos (opcional)
print("\nAvaliação da função nos pontos críticos:")
for pt in crit_points:
    val = f.subs(pt)
    print(f"{pt} -> f = {val}")
````

````python
import sympy as sp

# Variável simbólica
x = sp.Symbol('x', real=True)

# Função objetivo
f = x**2 + 1 / (x - 2)

# Gradiente (derivada)
grad = sp.diff(f, x)
print("Gradiente:")
sp.pprint(grad)

# Hessiana (segunda derivada)
hessiana = sp.diff(grad, x)
print("\nHessiana:")
sp.pprint(hessiana)

# Encontrar pontos críticos: f'(x) = 0
pontos_criticos = sp.solve(grad, x)
print("\nPontos críticos:")
for p in pontos_criticos:
    print(p)

# Verificar condições de otimalidade
for ponto in pontos_criticos:
    print(f"\nAnalisando ponto crítico x = {ponto}:")
    if ponto.is_real and ponto != 2:
        valor_hessiana = hessiana.subs(x, ponto).evalf()
        print(f" - Segunda derivada (aproximada): {valor_hessiana}")
        if valor_hessiana > 0:
            print(" Condição suficiente satisfeita: mínimo local.")
        elif valor_hessiana < 0:
            print(" Máximo local.")
        else:
            print(" Ponto de sela ou inflexão.")
    else:
        print(" Ponto não real ou fora do domínio (x ≠ 2).")
````

### Exercício 5.4

```math
\min_{x \in \mathbb{R}^2} x_1^6 - 3x_1^4 x_2^2 + 3x_1^2 x_2^4 - x_2^6


````python
import sympy as sp

# Variáveis simbólicas
x1, x2 = sp.symbols('x1 x2')
f = x1**6 - 3*x1**4*x2**2 + 3*x1**2*x2**4 - x2**6

# 1. Gradiente
grad_f = [sp.diff(f, var) for var in (x1, x2)]

# 2. Hessiana
hessian_f = sp.hessian(f, (x1, x2))

# 3. Resolver sistema: gradiente = 0
solutions = sp.solve(grad_f, (x1, x2), dict=True)

# 4. Classificar pontos críticos com base na Hessiana
def classify_point(sol):
    hessian_at_point = hessian_f.subs(sol)
    hessian_np = sp.Matrix(hessian_at_point).evalf()
    eigenvals = hessian_np.eigenvals()

    eigvals_list = list(eigenvals.keys())

    if all(ev > 0 for ev in eigvals_list):
        return "mínimo local"
    elif all(ev < 0 for ev in eigvals_list):
        return "máximo local"
    elif any(ev > 0 for ev in eigvals_list) and any(ev < 0 for ev in eigvals_list):
        return "ponto de sela"
    else:
        return "inconclusivo (caso degenerado)"

# Exibir resultados
print("Gradiente:")
print(grad_f)
print("\nHessiana:")
sp.pprint(hessian_f)

print("\nPontos críticos e classificação:")
for sol in solutions:
    tipo = classify_point(sol)
    print(f"Ponto: {sol}, tipo: {tipo}")
````

**Exercise 5.5.**

$$
\min_{x \in \mathbb{R}^2} f(x),
$$

onde \( f \) é definida por uma das funções do **Exercise 2.2**.


````python
import sympy as sp

# 1. Definição das variáveis
x1, x2 = sp.symbols('x1 x2')
f = x1**2 + x2**2  # Função estritamente convexa

# 2. Gradiente e Hessiana
grad_f = [sp.diff(f, var) for var in (x1, x2)]
hess_f = sp.hessian(f, (x1, x2))

# 3. Pontos críticos (onde gradiente = 0)
critical_points = sp.solve(grad_f, (x1, x2))

# 4. Análise da Hessiana no ponto crítico
hessian_at_crit = hess_f.subs({x1: critical_points[x1], x2: critical_points[x2]})
eigenvals = hessian_at_crit.eigenvals()
is_pos_def = all(val > 0 for val in eigenvals)

# Resultados
print("Função:", f)
print("Gradiente:", grad_f)
print("Hessiana:")
sp.pprint(hess_f)
print("Pontos críticos:", critical_points)
print("Hessiana no ponto crítico:")
sp.pprint(hessian_at_crit)
print("É positiva definida?", is_pos_def)
````
